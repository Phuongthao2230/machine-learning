**MODELING**

*RANDOM FOREST*

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
RFC = RandomForestClassifier(n_estimators=500)
```

```python
x = df.drop(columns=['Output', 'Reviews', 'Order Time','latitude','longitude'])
y = df['Output']
```

TÁCH TẬP DỮ LIỆU RA THÀNH TẬP TRAIN VÀ TẬP TEST

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100) 
```

```python
RFC.fit(X_train, y_train)
```

XÉT CÁC THUỘC TÍNH QUAN TRỌNG

```python
features_importance = np.array(RFC.feature_importances_)
features = np.array(x.columns)
```

```python
data={'features':features,'features_importance':features_importance}
fi_df = pd.DataFrame(data)
```

```python
fi_df.sort_values(by=['features_importance'], ascending=False,inplace=True)
sns.barplot(data=fi_df[:15], y='features', x='features_importance', color = "blue")
plt.show()
```

**Nhận xét:**
- Có thấy thể rằng feature "Ease and convenient", "Time saving" và "More restaurant choices" là những feature ảnh hưởng nhiều tới trong việc xác định "Output"

```python
RFC_pred = RFC.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, accuracy_score
```

ACCURACY

```python
print("Accuracy: ", round(accuracy_score(y_test, RFC_pred)*100,2))
```

BẢNG THỐNG KÊ PHÂN LỚP

```python
print(classification_report(y_test, RFC_pred))
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test, RFC_pred)
```

MA TRẬN NHẦM LẪN

```python
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
df_RF = pd.DataFrame(cf)
sn.set(font_scale=1.4) 
sn.heatmap(df_RF, annot=True, annot_kws={"size": 16}, fmt='d') 
plt.show()
```

*LOGISTIC REGRESSION*

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(max_iter=10000)
LR.fit(X_train, y_train)
LR_pred = LR.predict(X_test)
```

ACCURACY

```python
print("Accuracy", round(accuracy_score(LR_pred, y_test)*100,2))
```

BẢNG THỐNG KÊ PHÂN LỚP

```python
from sklearn.metrics import confusion_matrix
cf1 = confusion_matrix(y_test, LR_pred)
```

MA TRẬN NHẦM LẪN

```python
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
df_LR = pd.DataFrame(cf1)
sn.set(font_scale=1.4) 
sn.heatmap(df_LR, annot=True, annot_kws={"size": 16}, fmt='d') 
plt.show()
```

 *K NEAREST NEIGHBORS*

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

for i in range(1,11):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append([accuracy_score(y_test, pred_i), precision_score(y_test, pred_i), 
                       recall_score(y_test, pred_i)])

error_rate = pd.DataFrame(error_rate, columns=['accuracy_score', 'precision_score', 'recall_score'])
error_rate.plot()
```

**Nhận xét:**
Hiệu suất tốt nhất của mô hình KNN là khi k=7

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)

knn_pred = knn.predict(X_test)
```

ACCURACY

```python
print("Accuracy", round(accuracy_score(knn_pred, y_test)*100,2))
```

BẢNG THỐNG KÊ PHÂN LỚP

```python
print(classification_report(y_test, knn_pred))
```

```python
cf2 = confusion_matrix(y_test, knn_pred)
```

MA TRẬN NHẦM LẪN

```python
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
df_knn = pd.DataFrame(cf2)
sn.set(font_scale=1.4) 
sn.heatmap(df_knn, annot=True, annot_kws={"size": 16}, fmt='d') 
plt.show()
```
